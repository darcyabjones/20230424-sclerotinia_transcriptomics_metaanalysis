```python
dbcan = pd.read_csv("output/predector_protein/Scscl_1980-proteins-dbcan.tsv", sep="\t", na_values=".")
coverage = (dbcan["hmm_to"] - dbcan["hmm_from"]) / dbcan["hmm_len"]
is_long = (dbcan["query_to"] - dbcan["query_from"]) > 80
is_short = ~is_long

is_sig = (coverage > 0.3) & ((is_long & (dbcan["domain_i_evalue"] < 1e-5)) | (is_short & (dbcan["domain_i_evalue"] < 1e-3)))
dbcan = dbcan[is_sig]
dbcan
```

```python
rows = []

with open("work/dbcan_activities.txt") as handle:
    for line in handle:
        line = line.strip()
        if line.startswith("#"):
            continue

        line = line.split("\t", maxsplit=1)
        if len(line) == 1:
            #rows.append({"accession": line[0].strip(), "description": pd.NA})
            pass
        else:
            rows.append({"accession": line[0].strip(), "description": line[1].strip().replace("\t", " ")})

dbcan_desc = pd.DataFrame(rows).drop_duplicates()

dbcan_desc
```

```python
dbcan_tidy = dbcan["query hmm".split()].rename(columns={"query": "name", "hmm": "accession"})
dbcan_tidy["database"] = "dbCAN"
dbcan_tidy = dbcan_tidy["name database accession".split()]
dbcan_tidy = pd.merge(dbcan_tidy, dbcan_desc, on="accession", how="left")
dbcan_tidy
```

```python
phibase = pd.read_csv("output/predector_protein/Scscl_1980-proteins-phibase.tsv", sep="\t", na_values=".")
sig = (phibase["qcov"] > 0.3) & (phibase["tcov"] > 0.3) & (phibase["evalue"] < 1e-5)
phibase_tidy = phibase.loc[sig, "query target".split()]
tmp = phibase_tidy["target"].str.split("#")
phibase_tidy["phi"] = tmp.str[1]
phibase_tidy["phenotype"] = tmp.str[5]

phibase_tidy_phi = (
    phibase_tidy
    .drop(columns="phenotype")
    .set_index("query target".split())
    ["phi"].str.split("__")
    .apply(pd.Series)
    .stack()
    .reset_index()
    .drop(columns="level_2")
    .rename(columns={"query": "name", "target": "description", 0: "accession"})
    .drop_duplicates()
)
phibase_tidy_phi["database"] = "PHIbase"
phibase_tidy_phi = phibase_tidy_phi["name database accession description".split()]
phibase_tidy_phi

phibase_tidy_pheno = (
    phibase_tidy
    .drop(columns="phi")
    .set_index("query target".split())
    ["phenotype"].str.split("__")
    .apply(pd.Series)
    .stack()
    .reset_index()
    .drop(columns="level_2")
    .rename(columns={"query": "name", "target": "description", 0: "accession"})
    .drop_duplicates()
)
phibase_tidy_pheno["database"] = "PHIbasePhenotype"
phibase_tidy_pheno = phibase_tidy_pheno["name database accession description".split()]
phibase_tidy_pheno
```

```python
ips = pd.read_csv(
    "output/interproscan-protein.tsv",
    sep="\t",
    names="name md5 length analysis sig_acc sig_desc start stop score status date ipr_acc ipr_desc go pathways".split(),
    na_values="-"
)

ips
```

```python
gos = (
    ips[~ips["go"].isnull()]
    .set_index("name")
    ["go"]
    .str.split("|")
    .apply(pd.Series)
    .stack()
    .reset_index()
    .drop(columns="level_1")
    .rename(columns={0: "accession"})
    .drop_duplicates()
)

gos["database"] = "GO"
gos = gos["name database accession".split()]

def mapper(acc):
    g = goobo[acc]
    gd = str(g.definition)
    
    gn = str(g.name)
    return gn + "|" + gd

gos["description"] = gos["accession"].apply(mapper)

def mapper(acc):
    g = goobo[acc]
    if g.namespace == "molecular_function":
        gdd = "MF"
    elif g.namespace == "cellular_component":
        gdd = "CC"
    elif g.namespace == "biological_process":
        gdd = "BP"
    else:
        raise ValueError(g.namespace)
    return f"GO|{gdd}"

gos["database"] = gos["accession"].apply(mapper)

gos
```

```python
ipr = (
    ips[~ips["ipr_acc"].isnull()]
    ["name ipr_acc ipr_desc".split()]
    .rename(columns={"ipr_acc": "accession", "ipr_desc": "description"})
    .drop_duplicates()
    .reset_index(drop=True)
)

ipr["database"] = "interpro"
ipr = ipr["name database accession description".split()]
ipr.head()
```

```python
reactome_name_mapping = pd.read_csv("https://reactome.org/download/current/ReactomePathways.txt", sep="\t", names="accession description species".split())
reactome_name_mapping["description"] = reactome_name_mapping["description species".split()].apply(lambda x: f"{x.description} [{x.species}]", axis=1)
reactome_name_mapping.drop(columns="species", inplace=True)
reactome_name_mapping.head()
```

```python
pathways = (
    ips[~ips["pathways"].isnull()]
    .set_index("name")
    ["pathways"]
    .str.split("|")
    .apply(pd.Series)
    .stack()
    .reset_index()
    .drop(columns="level_1")
    .rename(columns={0: "accession"})
    .drop_duplicates()
)

tmp = pathways["accession"].str.split(":")
pathways["database"] = tmp.str[0].str.strip()
pathways["accession"] = tmp.str[1].str.strip()
pathways = pathways["name database accession".split()]
pathways
```

```python
pathways = pd.merge(pathways, reactome_name_mapping, on="accession", how="left")
```

```python
ips_uni = (
    ips
    ["name analysis sig_acc sig_desc".split()]
    .rename(columns={"analysis": "database", "sig_acc": "accession", "sig_desc": "description"})
    .drop_duplicates()
)

terms = pd.concat(
    [ips_uni, pathways, ipr, gos, dbcan_tidy, phibase_tidy_phi, phibase_tidy_pheno]
).drop_duplicates()
terms.sort_values(["name", "database", "accession"], inplace=True)
terms
```

```python
terms.to_csv("output/protein_functions_long.tsv", sep="\t", index=False, na_rep=".")
```

```python
swissprot = pd.read_csv("output/swissprot-matches.tsv", sep="\t")
swissprot
```

```python
swissprot_top5 = (
    swissprot
    .sort_values("query bits".split(), ascending=[True, False])
    .groupby("query")
    .head(5)
    .groupby("query")
    .apply(lambda df: pd.Series({"swissprot_protein_name": "|".join(df["protein_name"].unique()), "swissprot_gene_name": "|".join(df.loc[~df["gene_name"].isna(), "gene_name"].unique()),  "swissprot_accession": "|".join(df["accession"].unique())}))
    .reset_index()
    .rename(columns={"query": "name"})
)
swissprot_top5
```

```python
pdb = pd.read_csv("output/pdb-matches.tsv", sep="\t")
pdb
```

```python
pdb_top5 = (
    pdb
    .sort_values("query bits".split(), ascending=[True, False])
    .groupby("query")
    .head(5)
    .groupby("query")
    .apply(lambda df: pd.Series({"pdb_protein_name": "|".join(df["description"].unique()), "pdb_accession": "|".join(df["target"].unique())}))
    .reset_index()
    .rename(columns={"query": "name"})
)
pdb_top5
```

```python
def mapper(gos):
    goids = "|".join(gos)
    gonames = [goobo[g].name for g in gos]
    gonames = "|".join([g for g in gonames if not pd.isnull(g)])
    return pd.Series({"go_accession": goids, "go_names": gonames})

go_merged = gos.groupby("name")["accession"].apply(mapper).reset_index().pivot(index="name", columns="level_1").reset_index()
go_merged.index.name = None
go_merged.columns = ["name", "go_accession", "go_names"]
go_merged
```

```python
ipr_merged = ipr.groupby("name").apply(lambda x: pd.Series({"ipr_accession": "|".join([i for i in x.accession if not pd.isnull(i)]), "ipr_name": "|".join([i for i in x.description.unique() if not pd.isnull(i)])})).reset_index()
ipr_merged
```

```python
dbcan_merged = dbcan_tidy.groupby("name").apply(lambda x: pd.Series({"dbcan_accession": "|".join(x["accession"]), "dbcan_description": "|".join([list(set(i.split(";")))[0] for i in x["description"] if not pd.isnull(i)])})).reset_index()
```

```python
top5 = pd.merge(swissprot_top5, pdb_top5)
top5 = pd.merge(top5, go_merged)
top5 = pd.merge(top5, ipr_merged)
top5 = pd.merge(top5, dbcan_merged)
top5
```

```python
preds_merged = pd.merge(pred, top5, on="name", how="outer")
```
